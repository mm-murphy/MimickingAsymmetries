```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load in spreadsheet of results
rfile = './results.csv'
results = np.loadtxt(rfile, dtype=str, delimiter=',', skiprows=5, max_rows=13, unpack=True)
```

```python
# asymmetry factors
fasym = results[0]
# boundings
N1_bound = results[1].astype(float)[0]
N25_bound = results[7].astype(float)[0]
N50_bound = results[13].astype(float)[0]
N100_bound = results[19].astype(float)[0]
N200_bound = results[25].astype(float)[0]
N500_bound = results[31].astype(float)[0]
N1k_bound = results[37].astype(float)[0]
N2k_bound = results[43].astype(float)[0]
N10k_bound = results[49].astype(float)[0]
# fit t0 differences [minutes]
N1_tdiff = results[2].astype(float)
N25_tdiff = results[8].astype(float)
N50_tdiff = results[14].astype(float)
N100_tdiff = results[20].astype(float)
N200_tdiff = results[26].astype(float)
N500_tdiff = results[32].astype(float)
N1k_tdiff = results[38].astype(float)
N2k_tdiff = results[44].astype(float)
N10k_tdiff = results[50].astype(float)
# max absolute value in-transit data - homog. model residuals
N1_ITR = results[4].astype(float)
N25_ITR = results[10].astype(float)
N50_ITR = results[16].astype(float)
N100_ITR = results[22].astype(float)
N200_ITR = results[28].astype(float)
N500_ITR = results[34].astype(float)
N1k_ITR = results[40].astype(float)
N2k_ITR = results[46].astype(float)
N10k_ITR = results[52].astype(float)
# max true asym. model - b.f. homog. model residual
N1_MMR = results[5].astype(float)
N25_MMR = results[11].astype(float)
N50_MMR = results[17].astype(float)
N100_MMR = results[23].astype(float)
N200_MMR = results[29].astype(float)
N500_MMR = results[35].astype(float)
N1k_MMR = results[41].astype(float)
N2k_MMR = results[47].astype(float)
N10k_MMR = results[53].astype(float)
# by-eye minimum f_asym detectable
N1_mineye = 15
N25_mineye = 15
N50_mineye = 15
N100_mineye = 15
N200_mineye = 20
N500_mineye = 25
N1k_mineye = 35
N2k_mineye = 45
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(fasym, N1_tdiff, marker='o', label=str(N1_bound)+' min bound')
ax.plot(fasym, N25_tdiff, marker='o', label=str(N25_bound)+' min bound')
ax.plot(fasym, N50_tdiff, marker='o', label=str(N50_bound)+' min bound')
ax.plot(fasym, N100_tdiff, marker='o', label=str(N100_bound)+' min bound')
ax.plot(fasym, N200_tdiff, marker='o', label=str(N200_bound)+' min bound')
ax.plot(fasym, N500_tdiff, marker='o', label=str(N500_bound)+' min bound')
ax.plot(fasym, N1k_tdiff, marker='o', label=str(N1k_bound)+' min bound')
ax.plot(fasym, N2k_tdiff, marker='o', label=str(N2k_bound)+' min bound')
ax.plot(fasym, N10k_tdiff, marker='o', label=str(N10k_bound)+' min bound')
ax.set(ylim=(0, -2.))
ax.set(xlabel=r'$\Delta R_p / R_s$ [scale heights]', ylabel=r'$\Delta t_0$ [minute]')
ax.legend(loc='upper left', fontsize=8)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(fasym, N1_ITR, marker='o', label=str(N1_bound)+' min bound')
ax.plot(fasym, N25_ITR, marker='o', label=str(N25_bound)+' min bound')
ax.plot(fasym, N50_ITR, marker='o', label=str(N50_bound)+' min bound')
ax.plot(fasym, N100_ITR, marker='o', label=str(N100_bound)+' min bound')
ax.plot(fasym, N200_ITR, marker='o', label=str(N200_bound)+' min bound')
ax.plot(fasym, N500_ITR, marker='o', label=str(N500_bound)+' min bound')
ax.plot(fasym, N1k_ITR, marker='o', label=str(N1k_bound)+' min bound')
ax.plot(fasym, N2k_ITR, marker='o', label=str(N2k_bound)+' min bound')
ax.plot(fasym, N10k_ITR, marker='o', label=str(N10k_bound)+' min bound')
ax.axhline(50., c='black')
ax.set(xlabel=r'$\Delta R_p / R_s$ [scale heights]', ylabel=r'Max. Abs. I.T. D-H Residual [ppm]')
ax.legend(loc='upper left', fontsize=8)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(fasym, N1_MMR, marker='o', label=str(N1_bound)+' min bound')
ax.plot(fasym, N25_MMR, marker='o', label=str(N25_bound)+' min bound')
ax.plot(fasym, N50_MMR, marker='o', label=str(N50_bound)+' min bound')
ax.plot(fasym, N100_MMR, marker='o', label=str(N100_bound)+' min bound')
ax.plot(fasym, N200_MMR, marker='o', label=str(N200_bound)+' min bound')
ax.plot(fasym, N500_MMR, marker='o', label=str(N500_bound)+' min bound')
ax.plot(fasym, N1k_MMR, marker='o', label=str(N1k_bound)+' min bound')
ax.plot(fasym, N2k_MMR, marker='o', label=str(N2k_bound)+' min bound')
ax.plot(fasym, N10k_MMR, marker='o', label=str(N10k_bound)+' min bound')
# ax.axhline(50., c='black')
ax.set(ylim=(0,200))
ax.set(xlabel=r'$\Delta R_p / R_s$ [scale heights]', ylabel=r'Max. True - BF Model Residual [ppm]')
ax.legend(loc='upper left', fontsize=8)
plt.show()
```
