```python
import numpy as np
import matplotlib.pyplot as plt
```

This notebook aims to explore how one can improve the transit time uncertainty on a future observation by making intervening transit observations of some cadence and number. I follow the analytic prescriptions of Carter et. al 2008 for uncertainties on the transit time in terms of other relevant quantities.

```python
## relevant quantities
# variable definitions:
# Rp/Rs = planet-star radius ratio
# T14 = full transit duration
# b = impact parameter
# sigma = relative flux uncertainty on each lightcurve point
# c = observing cadence in [time units]
# nu = observing frequency (i.e. 1 / c)

def delta(RpRs):
    """ Computes the transit depth: delta = (Rp/Rs)^2, when input Rp/Rs"""
    return (RpRs**2)

def theta(RpRs, b):
    """ Computes the ratio of the ingress/egress duration, and the transit FWHM duration.
    Following Carter et. al 2008, we solve this in terms of the radius ratio and impact parameter
    theta = (Rp/Rs) / (1 - b^2)
    """
    return (RpRs)/(1. - b**2)

def Q(nu, T, depth, sigma):
    """ Computes approximate SNR of out-of-transit relative flux
    T ~ full transit duration
    nu ~ observing cadence frequency (units must match T)
    depth = transit depth
    sigma = photometric precision (same units as depth)
    """
    return (depth/sigma)*np.sqrt(nu*T)

def Tc_uncertainty(T, Q, theta):
    """ computes uncertainty on the transit centre time, using Carter et. al's equation
    returns in units of T
    """
    return (T/Q)*np.sqrt(0.5*theta)

def propagate_Tcerr(tpreverr, Ntransits, Perr):
    return np.sqrt((tpreverr**2) + (Ntransits**2)*(Perr**2))
```

The situation:

You have some existing ephemeris ($t_c \pm \sigma_{t_c}$; $P \pm \sigma_P$) from a previous transit. You wish to observe some future transit at $t_{new}$, by which the existing uncertainties on your ephemerides grow to an uncertainty on the future transit time of $\sigma_{t_{new}}$. In this future transit observation, you wish to investigate limb asymmetries. You are aware of the strong degeneracy between the effect of limb asymmetry and shifting of the transit time. To counter this, you wish to significantly reduce your uncertainty on the future transit time to a desired $\sigma_{t_{new}, ref}$, where $\sigma_{t_{new}, ref} < \sigma_{t_{new}}$. Here, we seek to explore how you can do this by observing some new transits in between. For now, let's assume you already know the period quite well.

This, of course, depends on how many transits you observe in between, what your target's magnitude is, and which instrument you're using (which determines the photometric precision). Let's simplify the latter by assuming you're using TESS observations. 

So, with N TESS transits and their prescribed cadence options, how does $\sigma_{t_{new}}$ map to $\sigma_{t_{new}, ref}$?

```python
## observation parameters
P = 10. # orbital period [day]
RpRs = 0.025
T14 = 4.0  # [hours] 
b = 0. # impact parameter
phot_precision = 1000. / 1.e6 # [ppm] -> converted to fractional
Ntesstransits = 3 # N intervening transits observed by TESS
```

```python
## for just one transit with TESS:
## not yet sure how to rope in Ntesstransits ...

Perr_vals = np.array([1.e-5, 1.e-6, 1.e-7, 1.e-8])
cad = 20. / (60.*60.) # cadence in [s] -> converted to [hours]
nu = 1./cad # [hour]^-1
depth = delta(RpRs)
Qval = Q(nu, T14, depth, phot_precision)
thetaval = theta(RpRs, b)
tc_TESSerr = Tc_uncertainty(T14, Qval, thetaval) # in [hours]
tc_TESSerr /= 24. # convert to [day]

## this may not be correct but:
tc_TESSerr /= np.sqrt(Ntesstransits)

tc_TESSerr_s = tc_TESSerr * 24. * 60. * 60.




Ntransitselapsing = np.linspace(1., 1000., 500) # N transits between last TESS transit and first pssible JWST transit
tc_JWSTerr_vals = np.zeros((len(Perr_vals), len(Ntransitselapsing)))
for i, Perr_val in enumerate(Perr_vals):
    for j, N in enumerate(Ntransitselapsing):
        tc_JWSTerr_vals[i,j] = propagate_Tcerr(tc_TESSerr, N, Perr_val) # in [day]
```

```python
fig, ax = plt.subplots(figsize=(8,4))
# colors = ['blue', ...]
ax.axhline(tc_TESSerr_s, c='black', alpha=0.2, label='TESS $\sigma_{t_c}$')
for i, Perr_size in enumerate(Perr_vals):
    ax.plot(Ntransitselapsing, tc_JWSTerr_vals[i]*24.*60.*60., label='$\sigma_P$ = '+str(Perr_vals[i]))
ax.set(xlabel='N transits between TESS and JWST', ylabel='Resulting precision for JWST observation [s]')
ax.legend(loc='best')
ax.set(xscale='log')
ax.set(ylim=(tc_TESSerr_s-5, tc_TESSerr_s+20))
plt.show()
```
