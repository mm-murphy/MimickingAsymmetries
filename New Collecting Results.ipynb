```python
extrafactor = 5./4.

## -- plotting all dt vs N
fig, ax = plt.subplots(figsize=(10,5))
# gj 1214
ax.text(0.85, 0.26, 'GJ-1214 b', c='coral', fontweight='bold', transform=ax.transAxes)
ax.errorbar(N_gj1214, abs(dtfits_gj1214), dterrs_gj1214, c='coral', marker='d', ls='None') 
ax.plot(N_gj1214, extrafactor*abs(0.5*dtcalc_gj1214), c='coral')
# wasp 80
ax.text(0.85, 0.15, 'WASP-80 b', c='green', fontweight='bold', transform=ax.transAxes)
ax.errorbar(N_w80, abs(dtfits_w80), dterrs_w80, c='green', marker='*', ls='None')
ax.plot(N_w80, extrafactor*abs(0.5*dtcalc_w80), c='green')
# wasp 96
ax.text(0.83, 0.48, 'WASP-96 b', c='purple', fontweight='bold', transform=ax.transAxes)
ax.errorbar(N_w96, abs(dtfits_w96), dterrs_w96, c='purple', marker='^', ls='None')
ax.plot(N_w96, extrafactor*abs(0.5*dtcalc_w96), c='purple')
# wasp 39
ax.text(0.83, 0.74, 'WASP-39 b', c='blue', fontweight='bold', transform=ax.transAxes)
ax.errorbar(N_w39, abs(dtfits_w39), dterrs_w39, c='blue', marker='s', ls='None')
ax.plot(N_w39, extrafactor*abs(0.5*dtcalc_w39), c='blue')
# wasp 107
ax.text(0.75, 0.89, 'WASP-107 b', c='black', fontweight='bold', transform=ax.transAxes)
ax.errorbar(N_w107, abs(dtfits_w107), dterrs_w107, c='black', marker='o', ls='None', label='Simulation Results')
ax.plot(N_w107, extrafactor*abs(0.5*dtcalc_w107), c='black', label=r'5/4 $\times$ Analytic Prediction')

#ax.set(xlabel='N', ylabel='dt [s]')
ax.set_xlabel('(Injected limb radius difference [N scale heights])\n'+r'$\bf{Smallest \ resolvable \ limb \ asymmetry \  [\Delta R_p \ in \ scale \ heights]}$', fontsize=11)
ax.set_ylabel(r'$\bf{Transit \ time \ uncertainty \ \sigma_{t_0} \ [s]}$'+'\n (|t$_{c, uniform~limb~best~fit}$ - t$_{c, true}$| [s])', fontsize=11)
ax.legend(loc='upper left', fontsize=10)
ax.xaxis.set_major_locator(MultipleLocator(1))
ax.xaxis.set_minor_locator(MultipleLocator(0.5))
ax.yaxis.set_major_locator(MultipleLocator(5))
ax.yaxis.set_minor_locator(MultipleLocator(2.5))
plt.savefig('analysis_figures/new_dt_vs_N_plot.png', dpi=400, bbox_inches='tight')
plt.show()
```
